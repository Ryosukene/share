```python
import torch
import numpy as np
import matplotlib.pyplot as plt
load_path = './mockingjay_model_tensor_165.pt'
load_weights2 = torch.load(load_path)
#net.load_state_dict(load_weights)
print(len(load_weights2))
print((load_weights2).shape)
#リニア値からdBへ変換
dBref = 2e-5
def db(x, dBref):
    y = 20 * np.log10(x / dBref)     #変換式
    return y                         #dB値を返す
img2=np.zeros((188,188))
for i in range(len(load_weights2[0][0])):
    for j in range(len(load_weights2[0][0])):
        img = load_weights2[0][0][i][j].to('cpu').detach().numpy().copy()
        img2[i][j]=db(img,dBref)
print(img2)
plt.imshow(img2)
plt.show()

```
